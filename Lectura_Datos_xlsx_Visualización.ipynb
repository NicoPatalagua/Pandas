La función rolling() de Pandas dataframe. proporciona la característica de cálculos de ventana rodante. El concepto de cálculo de la ventana rodante se utiliza principalmente en el procesamiento de señales y datos de series temporales. En palabras muy sencillas tomamos un tamaño de ventana de k a la vez y realizamos alguna operación matemática deseada en ella

```python
pres90days = fivePresidents.groupby('President', sort=False).rolling(1, on='End Date')['Approving'].mean()
pres90days.tail(14)
```

```python
pres50days = fivePresidents.groupby('President', sort=False).rolling('50D', on='End Date')['Approving'].mean()
pres50days.tail(14)
```

```python
styles = ['-.', '-', ':', '-', ':']
colors = [.9, .3, .7, .3, .9]
color = cm.Greys(colors)
title='90 Day Approval Rating Rolling Average'
plot_kwargs = dict(figsize=(16,6), style=styles,color = color, title=title)
correct_col_order = fivePresidents.President.unique()
pres90days.unstack('President')[correct_col_order].plot(**plot_kwargs)
 
```

# Visualización Con MatPlotLib: 


    
   1.- Se calcula la media del presupuesto para cada año, 
    y luego el promedio móvil de cinco años para suavizar los datos.


```python
movie = pd.read_csv('data/movie.csv')
med_budget = movie.groupby('title_year')['budget'].median() / 1e6
med_budget_roll = med_budget.rolling(5, min_periods=1).mean()
med_budget_roll.tail()
```

2.- Se llevan a Numpy Arrays

```python
years = med_budget_roll.index.values
years[-5:]
```

```python
budget = med_budget_roll.values
budget[-5:]
```


3.- Se grafica la mediana "rolling" de los presupuestos a lo largo del tiempo en una nueva figura.


```python
fig, ax = plt.subplots(figsize=(14,4), linewidth=5, edgecolor='.5')
ax.plot(years, budget, linestyle='--', linewidth=3, color='.2', label='Todas Las Películas')
text_kwargs=dict(fontsize=20, family='cursive')
ax.set_title('Presupuesto Medio por Película', **text_kwargs)
ax.set_ylabel('Milliones de Dólares', **text_kwargs)
```

Recuento de número de películas por año.

```python
movie_count = movie.groupby('title_year')['budget'].count()
movie_count.tail()
```

Como las unidades de ambos gráficos son completamente diferentes (dólares frente a recuento), se escalan los recuentos para que estén en el mismo rango que el presupuesto, adicional se etiqueta cada barra con su valor como texto. 
Como la gran mayoría de los datos están contenidos en los últimos años, se limitan los datos a las películas hechas a partir de 1970

```python
cuentat = movie_count.values
cuentat_norm = cuentat / cuentat.max() * budget.max()

cinco_year = (years % 5 == 0) & (years >= 1970)
years_5 = years[cinco_year]
ct_5 = cuentat[cinco_year]
ct_norm_5 = cuentat_norm[cinco_year]

ax.bar(years_5, ct_norm_5, 3, facecolor='.5', alpha=.3, label='Películas por Años')
ax.set_xlim(1968, 2017)
for x, y, v in zip(years_5, ct_norm_5, ct_5):
    ax.text(x, y + .5, str(v), ha='center')
ax.legend()
fig
```

A continuación un plot de dispersión del puntaje del IMDB contra el año para una selección aleatoria de 100 películas hechas a partir del año 2000. El tamaño de cada punto es proporcional al presupuesto.

```python
columnas = ['budget', 'title_year', 'imdb_score', 'movie_title']
nomissing = movie[columnas].dropna() #Se eliminan los valores NaN
nomissing['budget2'] = nomissing['budget'] / 1e6   # 1M
np.random.seed(0)  # Semilla para la aleatoriedad
muestra_movie_aleatoria = nomissing.query('title_year >= 2000').sample(100)

### Ahora se grafica 
fig, ax = plt.subplots(figsize=(14,6))
ax.scatter(x='title_year', y='imdb_score',s='budget2', data=muestra_movie_aleatoria)

idx_min = muestra_movie_aleatoria['imdb_score'].idxmin()
idx_max = muestra_movie_aleatoria['imdb_score'].idxmax()
for idx, offset in zip([idx_min, idx_max], [.5, -.5]):
    year = muestra_movie_aleatoria.loc[idx, 'title_year']
    score = muestra_movie_aleatoria.loc[idx, 'imdb_score']
    title = muestra_movie_aleatoria.loc[idx, 'movie_title']
    ax.annotate(xy=(year, score),
    xytext=(year + 1, score + offset),
    s=title + ' ({})'.format(score),
    ha='center',
    size=16,
    arrowprops=dict(arrowstyle="fancy"))
ax.set_title('Puntuación IMDB por Año', size=25)
ax.grid(True)
```

Analisis con Graficos de Vuelos (flights.csv)

```python
vuelos = pd.read_csv('data/flights.csv')
vuelos.head()
```

```python
vuelos['DELAYED'] = vuelos['ARR_DELAY'].ge(15).astype(int)
columnas = ['DIVERTED', 'CANCELLED', 'DELAYED']
vuelos['ON_TIME'] = 1 - vuelos[columnas].any(axis=1)

columnas.append('ON_TIME')
estado = vuelos[columnas].sum()
estado
```

Se grafica en multiples plots en la misma figura para las columnas categoricas y continuas.

```python
fig, ax_array = plt.subplots(2, 3, figsize=(18,8))
(ax1, ax2, ax3), (ax4, ax5, ax6) = ax_array
fig.suptitle('2015 USA Vuelos - Resumen Univariable', size=20)

aerolinea_cnt = vuelos['AIRLINE'].value_counts()
aerolinea_cnt.plot(kind='barh', ax=ax1, title ='Aerolínea')

org_cnt = vuelos['ORG_AIR'].value_counts()
org_cnt.plot(kind='bar', ax=ax2, rot=0, title='Ciudad de Origen')

dest_cnt = vuelos['DEST_AIR'].value_counts().head(10)
dest_cnt.plot(kind='bar', ax=ax3, rot=0, title='Ciudad de Destino')

estado.plot(kind='bar', ax=ax4, rot=0, log=True, title='Estado del Vuelo')
vuelos['DIST'].plot(kind='kde', ax=ax5, xlim=(0, 3000),
                     title='Distancia KDE')
vuelos['ARR_DELAY'].plot(kind='hist', ax=ax6, 
                          title='Retraso', range=(0,200))
```

Antes de pasar a las parcelas multivariantes, se grafica el número de vuelos por semana.  Infortunadamente, no tenemos el TimeStamp pandas en ninguna de las columnas, pero sí el mes y el día. La función to_datetime tiene un ingenioso truco que identifica los nombres de las columnas que coinciden con los componentes del TimeStamp.  Por ejemplo, si tienes un DataFrame con exactamente tres columnas año, mes y día, entonces al pasar este DataFrame a la función to_datetime se devolverá una secuencia de TimeStamps.  Para preparar nuestro actual DataFrame, necesitamos añadir una columna para el año y usar la hora de salida programada para obtener la hora y los minutos: 

```python
hora = vuelos['SCHED_DEP'] // 100
minuto = vuelos['SCHED_DEP'] % 100
df_fecha = vuelos[['MONTH', 'DAY']].assign(HOUR=hora, MINUTE=minuto, YEAR=2015)
df_fecha.head()
```

Podemos convertir este DataFrame en una Serie de TimeStamps con la función to_datetime.


```python
vuelos_sal = pd.to_datetime(df_fecha)
vuelos_sal.tail()
```

Usemos este resultado como nuestro nuevo índice y luego encontremos el conteo de vuelos por semana con el método de resample


```python
vuelos.index = vuelos_sal
vuelos_cnt = vuelos.resample('W').size()
vuelos_cnt.plot(figsize=(12,3), title='Vuelos Por Semana', grid=True)
```

```python
data_perdida = vuelos_cnt.where(vuelos_cnt >  1000)
vuelos_interp = data_perdida.interpolate(limit_direction='both')

ax = vuelos_interp.plot(color='black', figsize=(16,4))
vuelos_interp[vuelos_cnt < 500].plot(linewidth=10, grid=True, ax=ax)

ax.annotate(xy=(.8, .55), xytext=(.8, .77), 
            xycoords='axes fraction', s='Data Perdida', 
            ha='center',  size=20, arrowprops=dict())
ax.set_title('Vuelos por Semana (Data Perdida Interpolada)')
```

Encontrar los 10 aeropuertos que:
*Tienen la mayor distancia promedio recorrida para los vuelos de llegada
*Tienen un mínimo de 100 vuelos en total


```python
vuelos.groupby('DEST_AIR')['DIST'] \
       .agg(['mean', 'count']) \
       .query('count > 100') \
       .sort_values('mean') \
       .tail(10) \
       .plot(kind='bar', y='mean', legend=False, 
             rot=0, figsize=(14,4),
             title='Distancia Promedio por Destino')
```

Analicemos dos variables al mismo tiempo haciendo un gráfico de dispersión entre la distancia y el tiempo de vuelo para todos los vuelos de menos de 2.000 millas.

```python
vuelos_sc = vuelos.reset_index(drop=True)[['DIST', 'AIR_TIME']].query('DIST <= 2000').dropna()
vuelos_sc.plot(x='DIST', y='AIR_TIME', kind='scatter', s=1, figsize=(16,4))
```
